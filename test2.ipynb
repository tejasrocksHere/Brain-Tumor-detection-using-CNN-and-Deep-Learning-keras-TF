```python
from skimage import img_as_float
from skimage.metrics import peak_signal_noise_ratio
from matplotlib import pyplot as plt
from skimage import io
from scipy import ndimage as nd

# Read the noisy image
noisy_img = img_as_float(io.imread("pred\pred5.jpg"))

# Need to convert to float as we will be doing math on the array
# Also, most skimage functions need float numbers
ref_img = img_as_float(io.imread("pred\pred5.jpg"))

# Apply Gaussian filter
gaussian_img = nd.gaussian_filter(noisy_img, sigma=0.5)

# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(noisy_img, cmap='gray')
plt.title('Input Noisy Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gaussian_img, cmap='gray')
plt.title('Gaussian Filtered Image')
plt.axis('off')

plt.show()

# Save the denoised image
plt.imsave("m_new.jpg", gaussian_img, cmap='gray')

# Calculate PSNR
noise_psnr = peak_signal_noise_ratio(ref_img, noisy_img)
gaussian_cleaned_psnr = peak_signal_noise_ratio(ref_img, gaussian_img)

print("PSNR of input noisy image = ", noise_psnr)
print("PSNR of cleaned image = ", gaussian_cleaned_psnr)
```

```python
from scipy import ndimage

# Apply Sobel edge detection
sobel_img = ndimage.sobel(gaussian_img)
plt.imshow(sobel_img, cmap='gray')


# plt.imsave("m_new.jpg", gaussian_img, cmap='gray')
```
